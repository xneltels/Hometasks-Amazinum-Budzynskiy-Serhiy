```python
def get_drops(k, n):
    dp = [[0] * (n + 1) for _ in range(k + 1)]

    for i in range(1, k + 1):
        dp[i][1] = 1

    for j in range(1, n + 1):
        dp[1][j] = j

    for i in range(2, k + 1):
        for j in range(2, n + 1):
            dp[i][j] = 999999

            for x in range(1, j + 1):
                res = 1 + max(dp[i-1][x-1], dp[i][j-x])

                if res < dp[i][j]:
                    dp[i][j] = res

    return dp[k][n]

print("2 glasses, 100 floors:", get_drops(2, 100))
print("3 glasses, 1000 floors:", get_drops(3, 1000))
print("5 glasses, 1000 floors:", get_drops(5, 1000))
```
