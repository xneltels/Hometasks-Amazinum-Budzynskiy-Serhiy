```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

X, _ = make_classification(n_samples=200, n_features=2, n_informative=2,
                           n_redundant=0, random_state=0, n_clusters_per_class=1, class_sep=0.8)

k = 4
n_iters = 100
max_steps = 100

best_cost = float('inf')
best_labels = []
best_centroids = []

for attempt in range(n_iters):

    random_idx = np.random.choice(X.shape[0], k, replace=False)
    centroids = X[random_idx].copy()

    labels = np.zeros(X.shape[0])
    cost = 0

    for step in range(max_steps):
        for i in range(X.shape[0]):
            distances = []
            for j in range(k):
                dist = np.sqrt((X[i,0] - centroids[j,0])**2 + (X[i,1] - centroids[j,1])**2)
                distances.append(dist)
            labels[i] = np.argmin(distances)

        new_centroids = np.zeros((k, 2))
        for j in range(k):
            points_in_cluster = []
            for i in range(X.shape[0]):
                if labels[i] == j:
                    points_in_cluster.append(X[i])

            if len(points_in_cluster) > 0:
                new_centroids[j] = np.mean(points_in_cluster, axis=0)
            else:
                random_point = X[np.random.choice(X.shape[0])]
                new_centroids[j] = random_point

        if np.allclose(centroids, new_centroids):
            break

        centroids = new_centroids.copy()

    current_cost = 0
    for i in range(X.shape[0]):
        cluster_idx = int(labels[i])
        sq_dist = (X[i,0] - centroids[cluster_idx,0])**2 + (X[i,1] - centroids[cluster_idx,1])**2
        current_cost += sq_dist

    if current_cost < best_cost:
        best_cost = current_cost
        best_labels = labels.copy()
        best_centroids = centroids.copy()

plt.figure()
plt.scatter(X[:,0], X[:,1], c=best_labels)
plt.scatter(best_centroids[:,0], best_centroids[:,1], c='red', marker='x', s=100)
plt.title('K-Means clustering (best cost: ' + str(round(best_cost, 2)) + ')')
plt.show()
```
