```python
pairs = []
for a in range(2, 100):
    for b in range(a, 100):
        pairs.append((a, b))

def prod_knows(p):
    count = 0
    for a, b in pairs:
        if a * b == p:
            count += 1
    if count == 1:
        return True
    else:
        return False

def sum_knows(s):
    for a, b in pairs:
        if a + b == s:
            if prod_knows(a * b) == True:
                return False
    return True

def prod_knows_now(p):
    count = 0
    for a, b in pairs:
        if a * b == p:
            if sum_knows(a + b) == True:
                count += 1
    if count == 1:
        return True
    else:
        return False

def sum_knows_now(s):
    count = 0
    for a, b in pairs:
        if a + b == s:
            if prod_knows_now(a * b) == True:
                count += 1
    if count == 1:
        return True
    else:
        return False

def solve():
    for a, b in pairs:
        s = a + b
        p = a * b

        if prod_knows(p) == False:
            if sum_knows(s) == True:
                if prod_knows_now(p) == True:
                    if sum_knows_now(s) == True:
                        return a, b
    return None

ans = solve()
print(ans)
```

```python
from collections import defaultdict

def valid_pair(a, b):
    pairs = [(x, y) for x in range(1, 101) for y in range(x, 101)]

    sum_map = defaultdict(list)
    prod_map = defaultdict(list)

    for x, y in pairs:
        sum_map[x + y].append((x, y))
        prod_map[x * y].append((x, y))

    S = a + b
    P = a * b

    if len(sum_map[S]) <= 1:
        return False

    for x, y in prod_map[P]:
        if len(sum_map[x + y]) == 1:
            return False

    valid_products = []
    for x, y in prod_map[P]:
        s = x + y
        if len(sum_map[s]) > 1:
            valid_products.append((x, y))

    if len(valid_products) != 1:
        return False

    valid_sums = []
    for x, y in sum_map[S]:
        p = x * y
        valid_products = []

        for i, j in prod_map[p]:
            if len(sum_map[i + j]) > 1:
                valid_products.append((i, j))

        if len(valid_products) == 1:
            valid_sums.append((x, y))

    return len(valid_sums) == 1
```
